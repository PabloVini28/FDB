```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
import psycopg2 as pg
import sqlalchemy 
```

```python
con = pg.connect(host = 'localhost', dbname='sistema', user='postgres', password='28112003')
cursor = con.cursor()
cnx = 'postgresql://postgres:28112003@localhost/sistema'
sqlalchemy.create_engine(cnx)
```

```python
# Criando o output para exibição dos resultados
output = widgets.Output()

# Função para atualizar a exibição
def refresh_display(df=None):
    with output:
        output.clear_output()
        display(id_local,status,id_endereco, cep, rua, bairro, cidade, numero)
        # Create
        display(widgets.HBox([button_insert]))
        # Read All
        display(widgets.HBox([button_query_all]))
        # Read Local
        display(widgets.HBox([button_query_by_id_local,button_query_by_status]))
        # Read Endereco
        display(widgets.HBox([button_query_by_id_endereco,button_query_by_rua,button_query_by_bairro,button_query_by_cidade]))
        display(widgets.HBox([button_query_by_cep,button_query_by_numero]))
        # Update
        display(widgets.HBox([button_update_by_id]))
        # Delete
        display(widgets.HBox([button_delete_by_id]))
        
        if df is not None:
            display(df)

# Campos de entrada
id_local = widgets.IntText(value=0,description='ID Local:')
status = widgets.Dropdown( options=['True', 'False'], value='True', description='Status:')
cep = widgets.IntText(placeholder='Apenas números', description='CEP:')
rua = widgets.Text(placeholder='Rua Exemplo', description='Rua:')
bairro = widgets.Text(placeholder='Bairro Exemplo', description='Bairro:')
cidade = widgets.Text(placeholder='Cidade Exemplo', description='Cidade:')
numero = widgets.IntText(value=0,description='Numero:')
id_endereco = widgets.IntText(value=0,description='ID Endereco:')

# Botões
button_insert = widgets.Button(description='Inserir', button_style='success')
button_query_all = widgets.Button(description='Consultar Todos', button_style='info')
button_query_by_id_local = widgets.Button(description='Consultar por ID Local', button_style='info')
button_query_by_status = widgets.Button(description='Consultar por Status', button_style='info')
button_query_by_rua = widgets.Button(description='Consultar por Rua', button_style='info')
button_query_by_id_endereco = widgets.Button(description='Consultar por ID Endereço', button_style='info')
button_query_by_bairro = widgets.Button(description='Consultar por Bairro', button_style='info')
button_query_by_cidade = widgets.Button(description='Consultar por Cidade', button_style='info')
button_query_by_cep = widgets.Button(description='Consultar por CEP', button_style='info')
button_query_by_numero = widgets.Button(description='Consultar por Numero', button_style='info')
button_delete_by_id = widgets.Button(description='Deletar por ID', button_style='danger')
button_update_by_id = widgets.Button(description='Atualizar por ID', button_style='warning')

# Funções de ação

# Inserção
def on_button_clicked_insert(b):
    try:
        cursor.execute("INSERT INTO localizacao (status_verificado, endereco_id) VALUES (%s, %s)",(status.value,id_endereco.value))
        con.commit()
    except:
        cursor.execute("ROLLBACK")
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a inserção. Verifique restrições.")
    else:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id",cnx)
        refresh_display(df)

# Buscando Todos
def on_button_clicked_query_all(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id",cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.")

# Buscando por Status
def on_button_clicked_query_status(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE l.status_verificado = '{}'".format(status.value),cnx) 
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.")
            
# Buscando Por ID Local
def on_button_clicked_query_by_id_local(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE l.id = '{}'".format(id_local.value),cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.")

# Buscando por ID Endereco
def on_button_clicked_query_by_id_endereco(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE e.id = '{}'".format(id_endereco.value),cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.")

# Buscando por Rua
def on_button_clicked_query_by_rua(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE e.rua = '{}'".format(rua.value),cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.")

# Buscando por Cidade
def on_button_clicked_query_by_cidade(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE e.cidade = '{}'".format(cidade.value),cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.")

# Buscar por Bairro
def on_button_clicked_query_by_bairro(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE e.bairro = '{}'".format(bairro.value),cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.")

# Buscar por CEP
def on_button_clicked_query_by_cep(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE e.cep = '{}'".format(cep.value),cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.") 

# Buscar por Numero
def on_button_clicked_query_by_numero(b):
    try:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id WHERE e.numero = '{}'".format(numero.value),cnx)
        refresh_display(df)
    except:
        with output:
            output.clear_output()
            refresh_display("Não foi possível realizar a busca. Verifique restrições.") 

# Deletar por ID
def on_button_clicked_delete(b):
    try:
        cursor.execute("DELETE FROM localizacao WHERE id={}".format(id_local.value))
        con.commit()
    except:
        cursor.execute("ROLLBACK")
        with output:
            output.clear_output()
            display("Não foi possível realizar a exclusão. Verifique restrições.")
    else:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id",cnx)
        refresh_display(df)

# Atualizar por ID
def on_button_clicked_update(b):
    try:
        cursor.execute("UPDATE localizacao SET status_verificado = {0} WHERE id = {1}".format(status.value,id_local.value))
        con.commit()
    except:
        cursor.execute("ROLLBACK")
        with output:
            output.clear_output()
            display("Não foi possível realizar a atualização. Verifique restrições.")
    else:
        df = pd.read_sql("SELECT l.id AS localizacao_id, l.status_verificado, e.id AS endereco_id, e.cep, e.rua, e.bairro, e.cidade, e.numero FROM localizacao l JOIN endereco e ON l.endereco_id = e.id",cnx)
        refresh_display(df)
    
# Associando botões às funções
button_insert.on_click(on_button_clicked_insert) # inserir
button_query_all.on_click(on_button_clicked_query_all) # buscar todos
button_query_by_id_local.on_click(on_button_clicked_query_by_id_local) # buscar por id local
button_query_by_status.on_click(on_button_clicked_query_status) # buscar por status
button_query_by_id_endereco.on_click(on_button_clicked_query_by_id_endereco) # buscar por id endereco
button_query_by_rua.on_click(on_button_clicked_query_by_rua) # buscar por rua
button_query_by_cidade.on_click(on_button_clicked_query_by_cidade) # buscar por cidade
button_query_by_bairro.on_click(on_button_clicked_query_by_bairro) # buscar por bairro
button_query_by_cep.on_click(on_button_clicked_query_by_cep)# Buscar por CEP
button_query_by_numero.on_click(on_button_clicked_query_by_numero)# Buscar por Numero
button_delete_by_id.on_click(on_button_clicked_delete) # deletar
button_update_by_id.on_click(on_button_clicked_update) # atualizar

# Inicializa a interface
refresh_display()

# Exibe a saída
display(output)
```
